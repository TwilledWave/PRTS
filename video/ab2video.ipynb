```python
from voiceV import *

#generate video for the prts events
from pathlib import Path
import os

#lang zh, ja, en

def runab(query:str, folder = "./cache/unity/SL-ST/", lang = "en", overwrite = False, writeclip = True):
    import json, time
    stage_dict = {}; stage_list = [];
    # Opening JSON file
    with open(folder + "/"+ query, encoding="utf-8") as f:
        file = json.load(f)
    for l in file:
        if l != None:
            nameEN = l["stageEN"];
            nameCN = l["stageCN"];
            if lang =="zh":
                stage_list.append(nameCN);
                stage_dict[nameCN]=l["file"]
            if lang == "en":
                stage_list.append(nameEN);
                stage_dict[nameEN]=l["file"]
            if lang == "ja":
                stage_list.append(nameCN);
                stage_dict[nameCN]=l["file"]
    #open each stage
    print(stage_list);
    for k in range(len(stage_list)):
        stage = stage_list[k]
        link = folder + "/" + stage_dict[stage]
        link = 'file:///'+os.path.abspath(link)
        #create folder for audio and clips
        path = folder+"/clip/"+stage+"/"
        Path(os.path.dirname(path)).mkdir(parents=True, exist_ok=True)
        path = folder+"/audio/"+stage+"/"
        Path(os.path.dirname(path)).mkdir(parents=True, exist_ok=True)
        #generate video for each stage
        path = Path(folder+"/"+stage+".mp4")
        if not(path.is_file()) or (overwrite == True):
            print('running on stage '+stage+" "+link)
            link2video(link, folder = folder, stage = stage, overwrite = overwrite, writeclip = writeclip, lang = lang);
    return()
```

```python
#combine all stages into one video with title for each stage
def concat_all(query:str, folder = "./cache/unity/SL-ST/", lang = "en"):
    import moviepy.editor as mpy
    import time
    import json, time
    stage_dict = {}; stage_list = [];
    # Opening JSON file
    with open(folder + "/"+ query, encoding="utf-8") as f:
        file = json.load(f)
    for l in file:
        if l != None:
            nameEN = l["stageEN"];
            nameCN = l["stageCN"];
            if lang =="zh":
                stage_list.append(nameCN);
                stage_dict[nameCN]=l["file"]
            if lang == "en":
                stage_list.append(nameEN);
                stage_dict[nameEN]=l["file"]
            if lang == "ja":
                stage_list.append(nameCN);
                stage_dict[nameCN]=l["file"]
    clips = []; time_stamp = []; current_time = 0;
    for k in range(len(stage_list)):
        stage = stage_list[k]
        link = folder + "/" + stage_dict[stage]
        link = 'file:///'+os.path.abspath(link)
        path = folder+'/'+stage+".mp4"
        if Path(path).is_file():
            txt_clip = mpy.TextClip(stage, fontsize = 40, color = "white", font = "Microsoft-YaHei-&-Microsoft-YaHei-UI")
            txt_clip = txt_clip.set_pos('center').set_duration(1)
            back_clip = mpy.ColorClip(color = [0,0,0], size = (1024, 576)).set_duration(1)
            clip = mpy.CompositeVideoClip([back_clip, txt_clip])
            clip.fps = 12
            clips.append(clip)
            clips.append(mpy.VideoFileClip(path))
            current_time = current_time + 1 + mpy.VideoFileClip(path).duration
            time_stamp.append(stage + ": "+ time.strftime('%H:%M:%S', time.gmtime(current_time)))
    concat_clip = mpy.concatenate_videoclips(clips, method="chain")
    concat_clip.write_videofile(folder+'/'+query+".mp4", verbose=False, logger=None);
    return(time_stamp)
```

```python
time_stamp = concat_all("FD-ST.json", folder = "./cache/unity/IS4-JP/", lang = "en")
for t in time_stamp:
    print(t)
```
